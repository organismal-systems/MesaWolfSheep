# 🧮 A Mesa model of three trophically-linked populations
To set up this model, follow these four steps:
1. Set the parameters in the code box below:
     - the size of the grid is set by the width and height parameters
     - the size of the agents in the plot
         - psize: the size of the grass patch
         - marker: the size of wolves and sheep
         - 
    Size in the plots may need to be adjusted to be visually clear if you make the grid very large or small.
2. Execute `>> Restart the kernel and run all cells`.
    The first time you do this, check the `Do not ask again` box to speed things up later.
   Ignore the pink warning.
4. When the blue bar appears, click on the box at the right-hand side.
    The model is now ready to run.
## Running the model
To run a simulation:

5. Change parameters as appropriate for your investigation:
     - Set the random seed so that each run uses a different set of random decisions
     - Set the initial number of sheep and wolves
6. Click `RESET` and then the arrow, `>` to reinitialize the population and run the simulation.
7. Click `||` to pause the simulation, then restart or reset it.

## Changing grid size
8. To change the grid size, **repeat all the steps 1-7 above**.
